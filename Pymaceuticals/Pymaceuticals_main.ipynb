```python
 # Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "../../Homework/Instructions/Pymaceuticals/data/mouse_drug_data.csv"
clinical_trial_data_to_load = "../../Homework/Instructions/Pymaceuticals/data/clinicaltrial_data.csv"

# Create dataframes from the two CSV files
mouse_drug_data_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data_df = pd.read_csv(clinical_trial_data_to_load)
```

```python
# Merge the two dataframes on the mouse ID
mouse_clinic_df = pd.merge(mouse_drug_data_df, clinical_trial_data_df, on="Mouse ID")
mouse_clinic_df.head(5)
```

```python
# Tumor Volume Changes Over Time for Each Treatment

# Create a list of the drugs
# all_drugs = mouse_clinic_df['Drug'].unique()
drug_list = ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']

# Create a list of colors for the various plots
# colors = ["Red", "Blue", "Green", "Black", "gray", "Red", "Blue", "Green", "Black", "gray"]
colors = ["Red", "Blue", "Green", "Black"]

# Create a list of the marker styles for the various plots
makers = ["o","^","s","d"]

# Create a function to calculate and generate scatter plots for the tumor volume change over time for each treatment
def treatments(list_drugs):
    # Set the color and marker variables for the plots
    color = 0
    marker = 0
    
    # Loop through the list of drugs
    for drug in list_drugs:
        # Create a dataframe for each drug
        drug_df = mouse_clinic_df.loc[mouse_clinic_df['Drug'] == (drug)]
        
        # Create a list of the various time points within the dataframe
        times = drug_df['Timepoint'].unique()
        
        # Calculate the mean tumor volume for all mice at each timepoint
        means = [drug_df.loc[drug_df['Timepoint'] == (time)]['Tumor Volume (mm3)'].mean() for time in times]
        
        # Generate a sample by storing all tumor volumes at each timepoint in a list
        sample = [drug_df.loc[drug_df['Timepoint'] == (time)]['Tumor Volume (mm3)'] for time in times]
        
        # Set the X Axis to be each timepoint
        x_axis = drug_df['Timepoint'].unique()
        
        # Calculate the standard error
        standard_errors = [sem(s) for s in sample]
        
        # Create a plot based on the information for each drug
        plt.errorbar(x_axis, means, standard_errors, fmt=makers[marker],
                     color=colors[color], linestyle='--', linewidth=0.5)
        plt.grid(color='gray', linestyle='-', linewidth=0.2, axis='y')
        plt.title("Turmor Response to Treatment")
        plt.xlabel("Time (Days)")
        plt.ylabel("Tumor Volume (mm3)")
        plt.figlegend(drug_list, bbox_to_anchor=(0.23, 0.1, 0.1, 0.8))
        
        # Use the next color and marker for the next drug
        color +=1
        marker +=1
```

```python
# Display the plot for the tumor volume over time for each treatment
treatments(drug_list)
```

```python
# Metastatic Site Changes Over Time for Each Treatment

# Create a function to calculate and generate scatter plots for the number of met. sites over time for each treatment
def metastatic_sites(list_drugs):
    # Set the color variable for the plots
    color = 0
    marker = 0
    
    # Loop through the list of drugs
    for drug in list_drugs:
        # Create a dataframe for each drug
        drug_df = mouse_clinic_df.loc[mouse_clinic_df['Drug'] == (drug)]
        
        # Create a list of the various time points within the dataframe
        times = drug_df['Timepoint'].unique()
        
        # Calculate the mean number of met. sites for all mice at each timepoint
        means = [drug_df.loc[drug_df['Timepoint'] == (time)]['Metastatic Sites'].mean() for time in times]
        
        # Generate a sample by storing all met. sites at each timepoint in a list
        sample = [drug_df.loc[drug_df['Timepoint'] == (time)]['Metastatic Sites'] for time in times]
        
        # Set the X Axis to be each timepoint
        x_axis = drug_df['Timepoint'].unique()
        
        # Calculate the standard error
        standard_errors = [sem(s) for s in sample]
        
        # Create a plot based on the information for each drug
        plt.errorbar(x_axis, means, standard_errors, fmt=makers[marker],
                     color=colors[color], linestyle='--', linewidth=0.5)
        plt.grid(color='gray', linestyle='-', linewidth=0.2, axis='y')
        plt.title("Metastatic Spread During Treatment")
        plt.xlabel("Time (Days)")
        plt.ylabel("Metastatic Sites")
        plt.figlegend(drug_list, bbox_to_anchor=(0.23, 0.1, 0.1, 0.8))
        
        # Use the next color for the next graph
        color +=1
        marker +=1
```

```python
# Display the plots for the met. site changes over time for each treatment
metastatic_sites(drug_list)
```

```python
# Survival Rate During Treatment

# Create a function to calculate and generate scatter plots for the number of met. sites over time for each treatment
def survival(list_drugs):
    # Set the color variable for the plots
    color = 0
    marker = 0
    
    # Loop through the list of drugs
    for drug in list_drugs:
        # Create a dataframe for each drug
        drug_df = mouse_clinic_df.loc[mouse_clinic_df['Drug'] == (drug)]
        
        # Create a list of the various time points within the dataframe
        times = drug_df['Timepoint'].unique()
        
        # Calculate the total number of mice tested for each drug
        num_of_mice = drug_df['Mouse ID'].nunique()
        
        # Calculate the number of mice remaining at each timepoint
        num_of_mice_per_day = drug_df['Timepoint'].value_counts().tolist()

        # Calculate the survival percentage rate
        survival_rate = [(x / num_of_mice)*100 for x in num_of_mice_per_day]

        # Set the X Axis to be each timepoint
        x_axis = drug_df['Timepoint'].unique()
        

        # Create a plot based on the information for each drug
        plt.plot(x_axis, survival_rate, marker=makers[marker], color=colors[color], linestyle='--', linewidth=0.5)
        plt.grid(color='gray', linestyle='-', linewidth=0.2)
        plt.title("Survival During Treatment")
        plt.xlabel("Time (Days)")
        plt.ylabel("Survival Rate (%)")
        plt.ylim(31,105,10)
        plt.figlegend(drug_list, bbox_to_anchor=(0.24, 0.29, 0.1, 0.1))
        
        # Use the next color for the next graph
        color +=1
        marker +=1
```

```python
# Display the plots for the survival during treatment
survival(drug_list)
```

```python
# Tumor Change Over 45 Day Treatment

# Create a function to calculate and plot the tumor changes as percentages
def tumor_growth(list_drugs):
    
    # Create an empyt list to store the change percentages
    percentages = []
    
    # Set the bar graph value text postion
    x = -0.6
    
    # Loop through the list of drugs
    for drug in list_drugs:
        # Create a dataframe for each drug
        drug_df = mouse_clinic_df.loc[mouse_clinic_df['Drug'] == (drug)]
        
        # Create a list of the various time points within the dataframe
        times = drug_df['Timepoint'].unique()
        
        # Calculate the mean tumor volume for each time point of the drug
        means = [drug_df.loc[drug_df['Timepoint'] == (time)]['Tumor Volume (mm3)'].mean() for time in times]
        
        # Calculate the percentage tumor volume change
        percentage_change = ((means[(len(means)-1)]/means[0])*100)-100
        
        # Determine the color the drug should be represent with on the graph 
        if percentage_change >= 0:
            color = 'red'
            y = 5
        else:
            color = 'green'
            y = -8

        # Create a plot based on the information for each drug
        plt.bar(drug, percentage_change, color=(color), width= -1, align='edge')
        plt.grid(color='gray', linestyle='-', linewidth=0.5)
        plt.ylim(-30,70,20)
        plt.title("Tumor Change Over 45 Day Treatment")
        plt.ylabel("Percentage Tumor Volume Change")
        plt.text(x,y, s=(f"{(round(percentage_change))}%"), color='white')
        x += 1
```

```python
# Display the bar graph representing the tumer change over a 45 day treatment
tumor_growth(drug_list)
```
